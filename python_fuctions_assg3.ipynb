```python
# Ques 1: Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in
# the list.

# Ans 1:
def sum_of_even_numbers(numbers):
    return sum(num for num in numbers if num % 2 == 0)

# Example usage
numbers_list = [1, 2, 3, 4, 5, 6]
result = sum_of_even_numbers(numbers_list)
print("Sum of even numbers:", result)
```

```python
# Ques 3:  Implement a Python function that takes a list of integers and returns a new list containing the squares of
# each number.

# Ans 3:
def square_numbers(numbers):
    # Returning a new list with the squares of each number
    return [num ** 2 for num in numbers]

# Function to input numbers from the user
def input_numbers():
    # Taking the number of elements from the user
    n = int(input("How many numbers do you want to input? "))
    
    # empty list 
    numbers = []
    
    # Loop to take input for 'n' numbers
    for i in range(n):
        num = int(input(f"Enter number {i+1}: "))
        numbers.append(num)
    
    return numbers

squared_list = square_numbers(input_numbers())
print("List of squared numbers:", squared_list)
```

```python
# Ques 4: Write a Python function that checks if a given number is prime or not from 1 to 200.

# Ans 4:
def is_prime(num):
    if num < 2:
        return False
    for i in range(2, int(num**0.5) + 1):
        if num % i == 0:
            return False
    return True

# Taking input from the user
def check_prime():
    num = int(input("Enter a number between 1 and 200: "))
    
    if 1 <= num <= 200:
        if is_prime(num):
            print(f"{num} is a prime number.")
        else:
            print(f"{num} is not a prime number.")
    else:
        print("Please enter a number within the range of 1 to 200.")


check_prime()
```

```python
# Ques 5: Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of
# terms.

# Ans 5: 
class FibonacciIterator:
    def __init__(self, terms):
        self.terms = terms
        self.current = 0
        self.next_value = 1
        self.count = 0

    def __iter__(self):
        return self

    def __next__(self):
        if self.count < self.terms:
            fibonacci_number = self.current
            self.current, self.next_value = self.next_value, self.current + self.next_value
            self.count += 1
            return fibonacci_number
        else:
            raise StopIteration


num_terms = int(input("Enter the number of terms for the Fibonacci sequence: "))
fibonacci_sequence = FibonacciIterator(num_terms)

print("Fibonacci sequence:")
for number in fibonacci_sequence:
    print(number, end=' ')
```

```python
# Ques 7: Implement a generator function that reads a file line by line and yields each line as a string.

# Ans 7:
def read_file_line_by_line(file_path):
    try:
        with open(file_path, 'r') as file:
            for line in file:
                yield line.strip()  # Yield each line without trailing newline characters
    except FileNotFoundError:
        print(f"The file '{file_path}' was not found.")
    except Exception as e:
        print(f"An error occurred: {e}")


file_path = input("Enter the path of the file to read: ")

for line in read_file_line_by_line(file_path):
    print(line)
```

```python
# Ques 8:  Use a lambda function in Python to sort a list of tuples based on the second element of each tuple.

# Ans 8: 
# Sample list of tuples
tuples_list = [(1, 'banana'), (3, 'apple'), (2, 'cherry'), (4, 'date')]

# Sorting the list of tuples based on the second element
sorted_tuples = sorted(tuples_list, key=lambda x: x[1])

# Displaying the sorted list
print("Sorted list of tuples based on the second element:")
print(sorted_tuples)
```

```python
# Ques 9: Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit.

# Ans 9:
def celsius_to_fahrenheit(celsius):
    return (celsius * 9/5) + 32

# Sample list of temperatures in Celsius
celsius_temps = [0, 20, 37, 100]

# Using map() to convert the list of temperatures to Fahrenheit
fahrenheit_temps = list(map(celsius_to_fahrenheit, celsius_temps))

print("Celsius temperatures:", celsius_temps)
print("Fahrenheit temperatures:", fahrenheit_temps)
```

```python
# Ques 10: Create a Python program that uses `filter()` to remove all the vowels from a given string.

# Ans 10:
def remove_vowels(input_string):
    vowels = 'aeiouAEIOU'  # Define vowels (both lowercase and uppercase)

    # Using filter() to remove vowels from the input string
    filtered_string = ''.join(filter(lambda char: char not in vowels, input_string))
    
    return filtered_string

# Example usage
input_string = input("Enter a string: ")
result = remove_vowels(input_string)

print("String after removing vowels:", result)
```
